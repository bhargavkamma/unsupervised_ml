```python
# importing the necessary libraries

import matplotlib.pyplot as plt
import nltk
import string
import warnings
from scipy.stats import pearsonr
from nltk.corpus import stopwords
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download("stopwords")
warnings.filterwarnings('ignore')
```

```python
df.isnull().sum()
```

```python
splitted = df['posted'].str.split(' ', expand=True)

# create new columns
df['year'] = splitted[2].astype('int')
df['month'] = splitted[1]
```

```python
df['year'].value_counts().plot.bar()
plt.show()
```

```python
# remove irrelevant information

df['details'] = df['title'] + ' ' + df['details']

df = df[['main_speaker', 'details']]
df.dropna(inplace=True)
df.head()
```

```python
# make a copy of the data

data = df.copy()
```

```python
# text processing

def remove_stopwords(text):
    stop_words = stopwords.words('english')
    
    imp_words = []
    for word in str(text).split():
        word = word.lower()
        if word not in stop_words:
            imp_words.append(word)
    output = " ".join(imp_words)
    return output

```

```python
df['details'] = df['details'].apply(lambda text: remove_stopwords(text))

df.head()
```

```python
# remove punctuations

punctuations_list = string.punctuation

def cleaning_punc(text):
    signal = str.maketrans('', '', punctuations_list)
    return text.translate(signal)

df['details'] = df['details'].apply(lambda x: cleaning_punc(x))
df.head()
```

```python
# use wordcloud to visualize word occurence

details_corpus = " ".join(df['details'])

plt.figure(figsize=(20,20))
wc = WordCloud(max_words=1000,
               width=800,
               height=400).generate(details_corpus)

plt.axis('off')
plt.imshow(wc)
plt.show()
```

```python
# building the unsupervised machine learning recommendation system
vectorizer = TfidfVectorizer(analyzer = 'word')
vectorizer.fit(df['details'])
```

```python
df.head()
```

```python
# cosine similarity and pearson correalation

def get_similarities(talk_content, data=df):
    talk_array = vectorizer.transform(talk_content).toarray()
    sim = []
    pea = []
    for idx, row in data.iterrows():
        details = row['details']
        # get the vector for current talks
        talk_arrays = vectorizer.transform(
            data[data['details'] == details]['details']).toarray()
        # using as cosine sim
        cos_sim = cosine_similarity(talk_array, talk_arrays)[0][0]
        # pearson corr
        pea_sim = pearsonr(talk_array.squeeze(), talk_arrays.squeeze())[0]
        sim.append(cos_sim)
        pea.append(pea_sim)
    return sim,pea
```

```python
def recommend_talks(talk_content, data=data):
    data['cos_sim'], data['pea_sim'] = get_similarities(talk_content)
    data.sort_values(by=['cos_sim', 'pea_sim'], ascending=[False, False], inplace=True)
    display(data[['main_speaker', 'details']].head())
```

```python
talk_content = ['Time Management and working\
hard to become successful in life']

recommend_talks(talk_content)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Make sure 'details' column has no NaN values
df['details'] = df['details'].fillna("")

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['details'])

print("TF-IDF shape:", tfidf_matrix.shape)
```

```python
print(df.columns.tolist())
```

```python
def get_similar_talks(query, df, vectorizer, tfidf_matrix, top_k=5):
    """
    query: a text string (details or any text)
    df: DataFrame with at least 'details' (and optionally 'main_speaker')
    """
    from sklearn.metrics.pairwise import cosine_similarity
    import numpy as np

    q_vec = vectorizer.transform([query])
    sims = cosine_similarity(q_vec, tfidf_matrix).ravel()
    top_idx = np.argsort(-sims)[:top_k]

    cols_to_show = [c for c in ['details', 'main_speaker', 'cluster'] if c in df.columns]
    return df.iloc[top_idx][cols_to_show].assign(score=sims[top_idx])
```

```python
results = get_similar_talks("artificial intelligence in education", df, vectorizer, tfidf_matrix, top_k=5)
print(results)
```

```python
seed_idx = 10   # any row index you like
query_text = df.loc[seed_idx, "details"]

print("🔎 Seed talk details:\n", query_text, "\n")

results = get_similar_talks(query_text, df, vectorizer, tfidf_matrix, top_k=5)
print("✅ Recommended similar talks:\n")
print(results)
```
